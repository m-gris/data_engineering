```python
import pandas as pd 
import os, random
```

```python
# FUNCTION TO PICK_UP RANDOM FILE
# https://stackoverflow.com/questions/6411811/randomly-selecting-a-file-from-a-tree-of-directories-in-a-completely-fair-manner
def random_file(dir):
    files = [os.path.join(path, filename)
             for path, dirs, files in os.walk(dir)
             for filename in files
             if not filename.endswith(".bak")]

    return random.choice(files)
```

```python
path = random_file('data/log_data')
log_sample = pd.read_json(path, lines = True) # Read the file as a json object per line.
```
